```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fixed_alpha = pd.read_csv('fixed_alpha.txt')
fixed_alpha_no_lasso = pd.read_csv('fixed_alpha_no_lasso.txt')
decrease_alpha = pd.read_csv('alpha_decrease.txt')
```

```python
plt.plot(fixed_alpha['Epoch'], fixed_alpha['p3'],color='r', label='with Lasso')
plt.plot(fixed_alpha_no_lasso['Epoch'], fixed_alpha_no_lasso['p3'],color='b', label='no lasso' )
plt.plot(decrease_alpha['Epoch'], decrease_alpha['p3'],color='orange', label='decreasing alpha' )
plt.ylim(0.95,1)
plt.legend()
```

```python
plt.plot(fixed_alpha['Epoch'], fixed_alpha['p4'],color='r', label='with Lasso')
plt.plot(fixed_alpha_no_lasso['Epoch'], fixed_alpha_no_lasso['p4'],color='b', label='no lasso' )
plt.plot(decrease_alpha['Epoch'], decrease_alpha['p4'],color='orange', label='decreasing alpha' )
plt.ylim(-1,-0.95)
plt.legend()
```

```python
plt.plot(fixed_alpha['Epoch'], fixed_alpha['Loss_u'],color='r', label='with Lasso')
plt.plot(fixed_alpha_no_lasso['Epoch'], fixed_alpha_no_lasso['Loss_u'],color='b', label='no lasso' )
plt.plot(decrease_alpha['Epoch'], decrease_alpha['Loss_u'],color='orange', label='decreasing alpha')
plt.ylim(0,0.001)
plt.legend()
```

```python
plt.semilogy(fixed_alpha['Epoch'], fixed_alpha['Loss_pde'],color='r', label='with Lasso')
plt.semilogy(fixed_alpha_no_lasso['Epoch'], fixed_alpha_no_lasso['Loss_pde'],color='b', label='no lasso' )
plt.semilogy(decrease_alpha['Epoch'], decrease_alpha['Loss_pde'],color='orange', label='decreasing alpha')
plt.legend()
```

```python
plt.semilogy(fixed_alpha['Epoch'], fixed_alpha['Loss_tot'],color='r', label='with Lasso')
plt.semilogy(fixed_alpha_no_lasso['Epoch'], fixed_alpha_no_lasso['Loss_tot'],color='b', label='no lasso' )
plt.semilogy(decrease_alpha['Epoch'], decrease_alpha['Loss_tot'],color='orange', label='decreasing alpha')
plt.legend()
```
